```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
```

```python
# Initialize the driver
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 10)

# Navigate to the URL
url = "https://platform.stratascratch.com/coding?code_type=3&page_size=100&difficulties=3"
driver.get(url)

# Wait for page to load
time.sleep(5)
```

```python
data=[]

def extract_page_data():
    rows = driver.find_elements(By.CSS_SELECTOR, "a.group.contents")
    for row in rows:
        try:
            company = row.find_elements(By.CSS_SELECTOR, "div.flex.items-center")[1].text.strip()
            difficulty = row.find_element(By.CSS_SELECTOR, "span[class^='QuestionDifficulty']").text.strip()
            title = row.find_elements(By.CSS_SELECTOR, "div.flex.items-center")[4].text.strip()
            title = title.split('\n')[0] if '\n' in title else title
            link = row.get_attribute('href')
            
            title_with_link = f'=HYPERLINK("{link}","{title}")'
            
            data.append({
                'Company': company,
                'Difficulty': difficulty,
                'Title': title_with_link
            })
        except Exception as e:
            print(f"Error processing row: {e}")
            continue

# Loop through pages
for page in range(1, 3):  # This will get pages 1 and 2
    url = f"https://platform.stratascratch.com/coding?code_type=3&page_size=100&difficulties=3&page={page}"
    driver.get(url)
    time.sleep(5)  # Wait for page to load
    
    print(f"Extracting data from page {page}")
    extract_page_data()
```

```python
# Fourth cell: Create DataFrame
df = pd.DataFrame(data)  # This is the correct way
```
